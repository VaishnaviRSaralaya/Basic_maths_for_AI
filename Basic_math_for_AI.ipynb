Linear Algebra

Linear algebra is a branch of mathematics that deals with vectors, matrices, and linear transformations. Here, we'll demonstrate basic matrix operations, the concept of a determinant, and the computation of eigenvalues and eigenvectors.

1. Vectors
Vectors are quantities that have both magnitude and direction. They are represented mathematically in an ordered list format.

2. Matrices
Matrices are rectangular arrays of numbers arranged in rows and columns. They are used to organize data and perform calculations involving vectors.

   Some of the matrix operations are: Matrix Addition, Matrix Multiplication and Transpose of a Matrix

3. Linear Transformations
A linear transformation is a function that takes a vector as input and produces another vector as output, following specific rules. Linear transformations can change the vector's position, size, or orientation while keeping the operations linear. They can be represented mathematically as:

Eigenvalues and Eigenvectors
Eigenvalues and eigenvectors are key concepts in linear algebra, particularly in the context of linear transformations. An eigenvector of a matrix ùê¥  is a non-zero vector ùë£ such that when the matrix is applied to it, the output is a scalar multiple of the vector:

where ùúÜ is the corresponding eigenvalue. This means that the direction of the eigenvector remains unchanged when the transformation represented by matrix ùê¥ is applied.

Eigenvector: A vector that stays in the same direction when a matrix is applied.
Eigenvalue: A number that indicates how much the eigenvector is stretched or compressed.


```python
#  Linear Algebra:
#  Here I have solved some basic problems on matrix operations and finding eigenvalues and eigenvectors of a given matrix

import numpy as np

# Create matrices A and B
A = np.array([[10, 12], [18, -4]])
B = np.array([[76, 9], [21, 22]])

# Matrix addition
C = A + B
print("Matrix Addition:\n", C)

# Matrix multiplication
D = np.dot(A, B)
print("Matrix Multiplication:\n", D)

# Matrix transpose
E = A.T
print("Transpose of A:\n", E)

# Determinant of a matrix
det = np.linalg.det(B)
print("Determinant:", det)

# Eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

```python
# Linear Algebra: Solving a System of Equations
#This program solves a system of linear equations using NumPy.

import numpy as np
# Coefficients of the equations
# 2x + 3y = 8
# 4x + y = 7
A = np.array([[2, 3], [4, 1]])
B = np.array([8, 7])

# Solve for x and y
solution = np.linalg.solve(A, B)
print("Solution (x, y):", solution)
```

Differentiation:
Differentiation is the process of finding the rate at which something is changing. When you differentiate a function, you‚Äôre essentially finding its slope at any given point.
In math terms, if you have a function ùëì(ùë•), the derivative ùëì‚Ä≤(ùë•) gives you the slope of the tangent line to the curve at point ùë•.

Integration:
Integration is the process of finding the total accumulation of something. It helps you calculate areas under curves or total quantities. In math terms, if you have a function 
ùëì(ùë•), integration gives you the total area under the curve of the function between two points.

Differentiation: Finding how fast something is changing (slope).
Integration: Finding the total amount of something (area under the curve).

```python
# Finding the derivative and integral value of the given function

import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return 6*x**2 + 43*x - 52  # Example function

# Numerical differentiation using central difference method
def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

# Generate x values for plotting
x_vals = np.linspace(-10, 10, 200)
y_vals = f(x_vals)

# Compute derivative values
y_derivative = numerical_derivative(f, x_vals)

# Compute integral values using the cumulative trapezoidal rule
integral_vals = scipy.integrate.cumulative_trapezoid(y_vals, x_vals, initial=0)

# Find derivative at x=7
x_7 = 7
derivative_at_7 = numerical_derivative(f, x_7)
print(f"Derivative of f at x=7: {derivative_at_7}")

# Find definite integral from 0 to 6 using SciPy's quad function
integral_value, _ = scipy.integrate.quad(f, 0, 6)
print(f"Integral of f from 0 to 6: {integral_value}")

# Plot the function, its derivative, and its integral
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='f(x)', color='blue')
plt.plot(x_vals, y_derivative, label="f'(x) (Derivative)", linestyle='--', color='red')
plt.plot(x_vals, integral_vals, label="‚à´ f(x) dx (Integral)", linestyle='-.', color='green')

# Highlight x=7 for derivative
plt.scatter([x_7], [f(x_7)], color='black', marker='o', label='f(7)')
plt.scatter([x_7], [derivative_at_7], color='red', marker='x', label="f'(7)")

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function, Derivative, and Integral')
plt.legend()
plt.grid()
plt.show()
```

Probability
Probability is the study of how likely something is to happen. It helps us measure uncertainty and predict the likelihood of different events.

Probability is usually represented as a number between 0 and 1:
-> 0 means the event will never happen (impossible).

-> 1 means the event will definitely happen (certain).

-> A probability of 0.5 means there's an equal chance of the event happening or not happening.
For example, when flipping a fair coin:

The probability of getting heads is 0.5 (or 50%).
The probability of getting tails is also 0.5.

Key Points
Event: Any outcome or combination of outcomes (like rolling a die or drawing a card).

Sample Space: All possible outcomes of an event (for a die, it‚Äôs {1, 2, 3, 4, 5, 6}).

Calculating Probability: To find the probability of an event happening, you divide the number of favorable outcomes by the total number of outcomes.
In summary, probability helps us understand and quantify uncertainty by giving us a way to assess how likely different events are to occur. It‚Äôs widely used in games, statistics, science, and daily decision-making.

```python
# Probability
# 1. Tossing a coin

import numpy as np

# Probability of getting heads
p_heads = 0.5

# Number of times we flip the coin
num_flips = 100

# Simulate the coin flips
flips = np.random.rand(num_flips) < p_heads  # True = Heads, False = Tails

'''We Use np.random.rand(num_flips) < 0.5 to simulate coin flips.
It generates random numbers between 0 and 1.
If the number is less than 0.5, it's heads (True).
Otherwise, it's tails (False).'''

# Count heads and tails
heads_count = np.sum(flips)
tails_count = num_flips - heads_count

# Calculate probabilities
p_heads_simulated = heads_count / num_flips
p_tails_simulated = tails_count / num_flips

# Print results
print(f"Probability of Heads: {p_heads_simulated}")
print(f"Probability of Tails: {p_tails_simulated}")
```

```python
# Probability of Drawing a King from a Deck of Cards

from scipy.special import comb

# Total number of ways to pick 1 card from 52
total_ways = comb(52, 1)

# Favorable ways to pick a King (4 Kings in a deck)
favorable_ways = comb(4, 1)

# Probability
p_king = favorable_ways / total_ways

print(f"Probability of drawing a King: {p_king}")
```

Statistics:

Statistics is the science of collecting, analyzing, interpreting, and presenting data. It helps us understand and make sense of the information we gather.

Key Concepts in Statistics
Data Collection: The process of gathering information, such as through surveys, experiments, or observations. This data forms the basis for statistical analysis.

Descriptive Statistics: Techniques used to summarize and describe the main features of a dataset. 

Key measures include:
Mean: The average of a set of numbers, calculated by adding all values together and dividing by the number of values. It represents the central point of the data.

Mean=Sum¬†of¬†all¬†values/Number¬†of¬†values

Median: The middle value in a dataset when the numbers are arranged in order. If there is an even number of values, the median is the average of the two middle values. It helps understand the center of the data without being influenced by extreme values.

Mode: The value that appears most frequently in a dataset. There can be more than one mode (bimodal or multimodal) if multiple values appear with the same highest frequency.


Range: The difference between the highest and lowest values in a dataset. It provides a sense of how spread out the data is.
Range=Maximum¬†value‚àíMinimum¬†value

Inferential Statistics: Techniques used to make predictions or generalizations about a larger population based on a sample of data. This often involves:

Hypothesis Testing: A method of testing a claim about a population using sample data. It helps determine whether the evidence supports or rejects the claim.

Confidence Intervals: A range of values used to estimate a population parameter. It provides a way to express uncertainty about the estimate.

Probability: A key concept interwoven with statistics that deals with the likelihood of events occurring. It helps in making inferences about populations based on sample data.

Data Visualization: Representing data through charts, graphs, or tables to make it easier to understand patterns and trends. Common visualizations include bar charts, histograms, pie charts, and scatter plots.
